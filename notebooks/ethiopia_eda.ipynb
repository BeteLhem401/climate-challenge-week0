# Imports


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import zscore
```

# Load Dataset


```python
df = pd.read_csv("../data/ethiopia.csv")
df.head()
```

# Add Country Column


```python
df["Country"] = "Ethiopia"
```

# Create Date Column


```python
df["DATE"] = pd.to_datetime(
    df["YEAR"] * 1000 + df["DOY"],
    format="%Y%j"
)
```

# Extract Month


```python
df["Month"] = df["DATE"].dt.month
df.head()
```

# Replace Missing Values


```python
df.replace(-999, np.nan, inplace=True)
```

# Check Duplicates


```python
duplicates = df.duplicated().sum()
print("Duplicates:", duplicates)
```

# Remove Duplicates


```python
df = df.drop_duplicates()
```

# Summary Statistics


```python
df.describe()
```

# Missing Value Report


```python
missing = df.isna().sum()
missing_percent = (df.isna().sum() / len(df)) * 100

missing_report = pd.DataFrame({
    "Missing Count": missing,
    "Missing %": missing_percent
})

missing_report.sort_values("Missing %", ascending=False)
```

NO missing value


# Outlier Detection


```python
cols = [
    "T2M",
    "T2M_MAX",
    "T2M_MIN",
    "PRECTOTCORR",
    "RH2M",
    "WS2M",
    "WS2M_MAX"
]

z_scores = np.abs(df[cols].apply(zscore))

outliers = (z_scores > 3).sum()
print(outliers)
```

# Export Clean Dataset


```python
df.to_csv("../data/ethiopia_clean.csv", index=False)
```

# Monthly Average Temperature


```python
monthly_temp = df.groupby("Month")["T2M"].mean()

monthly_temp.plot(figsize=(10, 5))
plt.title("Monthly Average Temperature - Ethiopia")
plt.xlabel("Month")
plt.ylabel("Temperature")
plt.grid(True)
plt.show()
```

# Monthly Rainfall


```python
monthly_rain = df.groupby("Month")["PRECTOTCORR"].sum()

monthly_rain.plot(kind="bar", figsize=(10, 5))
plt.title("Monthly Rainfall - Ethiopia")
plt.xlabel("Month")
plt.ylabel("Rainfall")
plt.show()
```

# Correlation Heatmap


```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

# Scatter Plot Temperature vs Humidity


```python
sns.scatterplot(data=df, x="T2M", y="RH2M")
plt.title("Temperature vs Humidity")
plt.show()
```

# Scatter Plot Temperature Range vs Wind Speed


```python
sns.scatterplot(data=df, x="T2M_RANGE", y="WS2M")
plt.title("Temperature Range vs Wind Speed")
plt.show()
```

# Rainfall Histogram


```python
df["PRECTOTCORR"].hist(bins=30)
plt.title("Rainfall Distribution")
plt.xlabel("Rainfall")
plt.ylabel("Frequency")
plt.show()
```

# Bubble Chart


```python
plt.figure(figsize=(10, 6))

plt.scatter(
    df["T2M"],
    df["RH2M"],
    s=df["PRECTOTCORR"] * 5,
    alpha=0.5
)

plt.xlabel("Temperature")
plt.ylabel("Humidity")
plt.title("Temperature vs Humidity with Rainfall Bubble Size")
plt.show()
```
